## Problem Space

Saudi Arabia launched **Vision 2030** in April 2016 — an ambitious national transformation plan designed to reduce the country's dependence on oil, diversify the economy, and modernize society. A decade into implementation, it raises a compelling question:

> **Is Saudi Arabia actually transforming its economy, or is oil still king?**

This project uses World Bank development indicators spanning **1990 to 2023** to track Saudi Arabia's economic trajectory across key pillars: GDP growth, oil dependency, tourism, female workforce participation, education, and population trends. The goal is to tell a data-driven story about the Kingdom's transformation — or lack thereof.

**Why this matters:**
- Saudi Arabia is the world's largest oil exporter and a top-15 global economy
- Vision 2030 targets are widely watched by investors, policymakers, and regional governments
- Understanding whether diversification is real has implications for the entire Gulf region

**Stakeholders:** economists, investors, policymakers, Saudi citizens, regional governments, international businesses

---

## Dataset

**Source:** [World Bank Development Indicators](https://data.worldbank.org/country/saudi-arabia) — the primary global collection of development statistics compiled from officially recognized international sources.

**Coverage:** Saudi Arabia, 1990–2023 (33 years)

**Key indicators tracked:**
- GDP (current USD) and GDP growth rate
- Oil rents as % of GDP
- Tourism arrivals and receipts
- Female labor force participation rate
- Unemployment rate
- Population
- School enrollment (tertiary)
- CO₂ emissions per capita

The dataset connects directly to the problem: each indicator maps to a specific Vision 2030 pillar — economic diversification (GDP/oil), social reform (female employment), tourism growth, and sustainability (CO₂).

1. How has Saudi Arabia's GDP grown over the past three decades, and what does the trend look like since Vision 2030 launched (2016)?

## 4. Data Analysis & Answers to Questions

### Q1: How has Saudi Arabia's GDP grown over three decades?

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# GDP absolute
ax1 = axes[0]
colors = ['#1a7f4b' if e == 'Post-2016 (Vision 2030)' else '#2c5f8a'
          for e in df['vision_2030_era']]
ax1.bar(df['year'], df['gdp_billion_usd'], color=colors, width=0.8)
ax1.axvline(2016, color='red', linestyle='--', linewidth=1.5, label='Vision 2030 Launch')
ax1.set_title('Saudi Arabia GDP (Billion USD)', fontweight='bold')
ax1.set_xlabel('Year'); ax1.set_ylabel('Billion USD')
ax1.legend()

# GDP growth rate
ax2 = axes[1]
bar_colors = ['#e05c2a' if g < 0 else '#2c7be5' for g in df['gdp_growth_pct']]
ax2.bar(df['year'], df['gdp_growth_pct'], color=bar_colors, width=0.8)
ax2.axhline(0, color='black', linewidth=0.8)
ax2.axvline(2016, color='red', linestyle='--', linewidth=1.5, label='Vision 2030 Launch')
ax2.set_title('Annual GDP Growth Rate (%)', fontweight='bold')
ax2.set_xlabel('Year'); ax2.set_ylabel('Growth (%)')
ax2.legend()

plt.tight_layout()
plt.savefig('q1_gdp_growth.png', dpi=150, bbox_inches='tight')
plt.show()

gdp_2016 = df.loc[df['year']==2016, 'gdp_billion_usd'].values[0]
gdp_2023 = df.loc[df['year']==2023, 'gdp_billion_usd'].values[0]
print(f"\nGDP in 2016 (Vision 2030 launch): ${gdp_2016:.0f}B")
print(f"GDP in 2023: ${gdp_2023:.0f}B")
print(f"Growth since 2016: {((gdp_2023-gdp_2016)/gdp_2016)*100:.1f}%")
```

**Answer Q1:** Saudi Arabia's GDP has grown dramatically from ~$117B in 1990 to over $1T in 2022 — driven largely by oil price cycles. Post-2016 GDP hit a record $1.1 trillion in 2022 on the back of high oil prices, representing a 65% increase from the Vision 2030 launch year. The 2020 dip reflects the COVID-19 pandemic. Volatility persists, tightly tied to oil markets.

### Q2: Is Saudi Arabia becoming less dependent on oil?

```python
fig, ax = plt.subplots(figsize=(13, 5))

ax.fill_between(df['year'], df['oil_rents_pct_gdp'], alpha=0.3, color='#c49a00')
ax.plot(df['year'], df['oil_rents_pct_gdp'], color='#c49a00', linewidth=2.5, label='Oil Rents % of GDP')
ax.axvline(2016, color='red', linestyle='--', linewidth=1.5, label='Vision 2030 Launch')

# Annotate key points
peak_yr = df.loc[df['oil_rents_pct_gdp'].idxmax(), 'year']
peak_val = df['oil_rents_pct_gdp'].max()
ax.annotate(f'Peak: {peak_val:.1f}% ({peak_yr})',
            xy=(peak_yr, peak_val), xytext=(peak_yr-5, peak_val+2),
            arrowprops=dict(arrowstyle='->', color='black'), fontsize=10)

ax.set_title('Oil Rents as % of GDP — Is Saudi Arabia Diversifying?', fontweight='bold')
ax.set_xlabel('Year'); ax.set_ylabel('Oil Rents (% of GDP)')
ax.legend()
plt.savefig('q2_oil_dependency.png', dpi=150, bbox_inches='tight')
plt.show()

avg_pre = df.loc[df['year'] < 2016, 'oil_rents_pct_gdp'].mean()
avg_post = df.loc[df['year'] >= 2016, 'oil_rents_pct_gdp'].mean()
print(f"Average oil rents % — Pre-2016: {avg_pre:.1f}%")
print(f"Average oil rents % — Post-2016: {avg_post:.1f}%")
print(f"Change: {avg_post - avg_pre:+.1f} percentage points")
```

**Answer Q2:** Oil rents as a share of GDP have shown some decline post-2016 (avg ~23% vs ~29% pre-2016), though this partly reflects oil price fluctuations rather than pure diversification. The 2020 plunge to ~12% was COVID-driven (collapsed prices). True structural diversification remains a work in progress — non-oil sectors like tourism and finance are growing but oil revenue still drives the fiscal budget.

### Q3: How has female labor force participation changed?

```python
fig, ax = plt.subplots(figsize=(13, 5))

pre = df[df['year'] < 2016]
post = df[df['year'] >= 2016]

ax.plot(pre['year'], pre['female_lfp_pct'], 'o-', color='#6b8cba', linewidth=2, markersize=5, label='Pre-Vision 2030')
ax.plot(post['year'], post['female_lfp_pct'], 's-', color='#c0392b', linewidth=2.5, markersize=7, label='Vision 2030 Era')
ax.axvline(2016, color='gray', linestyle='--', linewidth=1.2)

# Highlight key reform years
ax.annotate('Driving rights granted\n(2018)', xy=(2018, 19.9),
            xytext=(2014, 22), arrowprops=dict(arrowstyle='->', color='black'), fontsize=9)

ax.set_title('Female Labor Force Participation Rate (%) — A Vision 2030 Milestone', fontweight='bold')
ax.set_xlabel('Year'); ax.set_ylabel('Female LFP (%)')
ax.set_ylim(10, 40)
ax.legend()
plt.savefig('q3_female_lfp.png', dpi=150, bbox_inches='tight')
plt.show()

lfp_2016 = df.loc[df['year']==2016, 'female_lfp_pct'].values[0]
lfp_2023 = df.loc[df['year']==2023, 'female_lfp_pct'].values[0]
print(f"Female LFP in 2016: {lfp_2016}%")
print(f"Female LFP in 2023: {lfp_2023}%")
print(f"Absolute increase since Vision 2030: +{lfp_2023 - lfp_2016:.1f} percentage points")
print(f"That's a {((lfp_2023-lfp_2016)/lfp_2016)*100:.0f}% relative increase!")
```

**Answer Q3:** This is one of Vision 2030's clearest success stories. Female labor force participation surged from ~17% in 2016 to over 34% by 2023 — nearly doubling in just 7 years. The acceleration is especially visible after 2018, coinciding with major social reforms including women gaining the right to drive, attend sports events, and access more employment sectors. This is the most dramatic structural shift visible in the data.

### Q4: What is the trend in tourism arrivals?

```python
fig, ax = plt.subplots(figsize=(13, 5))

ax.fill_between(df['year'], df['tourism_arrivals_millions'], alpha=0.25, color='#27ae60')
ax.plot(df['year'], df['tourism_arrivals_millions'], 'o-', color='#27ae60', linewidth=2.5, markersize=5)
ax.axvline(2016, color='red', linestyle='--', linewidth=1.5, label='Vision 2030 Launch')

# COVID annotation
ax.annotate('COVID-19\n(2020)', xy=(2020, 1.7), xytext=(2017, 5),
            arrowprops=dict(arrowstyle='->', color='darkred', lw=1.5),
            fontsize=10, color='darkred')

ax.annotate('Record: 24.9M\n(2023)', xy=(2023, 24.9), xytext=(2019, 22),
            arrowprops=dict(arrowstyle='->', color='darkgreen'), fontsize=10, color='darkgreen')

ax.set_title('International Tourism Arrivals to Saudi Arabia (Millions)', fontweight='bold')
ax.set_xlabel('Year'); ax.set_ylabel('Arrivals (Millions)')
ax.legend()
plt.savefig('q4_tourism.png', dpi=150, bbox_inches='tight')
plt.show()

print("Tourism arrivals by era:")
print(df.groupby('vision_2030_era')['tourism_arrivals_millions'].agg(['mean','max','min']).round(1))
```

**Answer Q4:** Tourism arrivals reached a record **24.9 million in 2023**, up from ~16M at Vision 2030's launch. Much of the historical growth was driven by Hajj/Umrah religious pilgrims, but post-2016 investments (NEOM, Red Sea resorts, entertainment licenses) are pulling in leisure travelers. The 2020 collapse to 1.7M was entirely pandemic-related. The recovery and new record signals real momentum — Vision 2030's target of 100M visitors by 2030 remains ambitious but directionally supported.

### Q5: How has the unemployment rate evolved?

```python
fig, ax = plt.subplots(figsize=(13, 5))

ax.plot(df['year'], df['unemployment_pct'], 's-', color='#8e44ad', linewidth=2.5, markersize=6)
ax.axvline(2016, color='red', linestyle='--', linewidth=1.5, label='Vision 2030 Launch')
ax.axhline(df['unemployment_pct'].mean(), color='gray', linestyle=':', linewidth=1.2,
           label=f"Overall mean: {df['unemployment_pct'].mean():.1f}%")

ax.annotate('COVID-19 spike', xy=(2020, 7.4), xytext=(2017, 7.8),
            arrowprops=dict(arrowstyle='->', color='darkred'), fontsize=10)

ax.set_title('Total Unemployment Rate (%) — Saudi Arabia', fontweight='bold')
ax.set_xlabel('Year'); ax.set_ylabel('Unemployment (%)')
ax.set_ylim(4, 10)
ax.legend()
plt.savefig('q5_unemployment.png', dpi=150, bbox_inches='tight')
plt.show()

print(f"Unemployment 2016: {df.loc[df['year']==2016,'unemployment_pct'].values[0]}%")
print(f"Unemployment 2023: {df.loc[df['year']==2023,'unemployment_pct'].values[0]}%")
print(f"Peak unemployment: {df['unemployment_pct'].max()}% in {df.loc[df['unemployment_pct'].idxmax(),'year']}")
```

**Answer Q5:** Unemployment stood at a relatively low 5.8% in 2023, recovering from a COVID peak of 7.4% in 2020. The trend is generally stable but doesn't show dramatic improvement — notable because Vision 2030 places heavy emphasis on Saudization (replacing expat workers with Saudi nationals) which can temporarily push unemployment metrics up before they settle.

### Q6: Does GDP rise and fall with oil prices? (Correlation)

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Scatter: oil rents vs GDP
ax1 = axes[0]
scatter_colors = ['#e74c3c' if e == 'Post-2016 (Vision 2030)' else '#3498db'
                  for e in df['vision_2030_era']]
ax1.scatter(df['oil_rents_pct_gdp'], df['gdp_billion_usd'], c=scatter_colors, s=60, alpha=0.8)
# Trend line
z = np.polyfit(df['oil_rents_pct_gdp'], df['gdp_billion_usd'], 1)
p = np.poly1d(z)
x_range = np.linspace(df['oil_rents_pct_gdp'].min(), df['oil_rents_pct_gdp'].max(), 100)
ax1.plot(x_range, p(x_range), 'k--', linewidth=1.5, alpha=0.6)
ax1.set_xlabel('Oil Rents (% of GDP)'); ax1.set_ylabel('GDP (Billion USD)')
ax1.set_title('Oil Rents vs GDP', fontweight='bold')
from matplotlib.patches import Patch
legend_elements = [Patch(facecolor='#3498db', label='Pre-2016'),
                   Patch(facecolor='#e74c3c', label='Post-2016')]
ax1.legend(handles=legend_elements)

# Oil vs GDP over time (dual axis)
ax2 = axes[1]
ax2b = ax2.twinx()
ax2.plot(df['year'], df['gdp_billion_usd'], 'b-', linewidth=2, label='GDP (B USD)')
ax2b.plot(df['year'], df['oil_rents_pct_gdp'], 'orange', linewidth=2, linestyle='--', label='Oil Rents %')
ax2.set_xlabel('Year'); ax2.set_ylabel('GDP (Billion USD)', color='blue')
ax2b.set_ylabel('Oil Rents (% of GDP)', color='orange')
ax2.set_title('GDP & Oil Rents Over Time', fontweight='bold')

plt.tight_layout()
plt.savefig('q6_oil_gdp_correlation.png', dpi=150, bbox_inches='tight')
plt.show()

corr = df['oil_rents_pct_gdp'].corr(df['gdp_growth_pct'])
print(f"Correlation between oil rents % and GDP growth rate: {corr:.3f}")
print(f"Interpretation: {'Positive — GDP tends to grow when oil share is high' if corr > 0 else 'Mixed relationship'}")
```

**Answer Q6:** The scatter plot and dual-axis chart confirm that Saudi GDP is still heavily correlated with oil. Years of high oil rents tend to coincide with GDP surges (2008, 2011-2014, 2022). The positive correlation coefficient confirms GDP growth tracks oil conditions. The post-2016 red points (Vision 2030 era) show the Kingdom hasn't yet broken this oil-GDP linkage — GDP booms like 2022 are still oil-price-driven.

### Q7: How have CO₂ emissions per capita changed?

```python
fig, ax = plt.subplots(figsize=(13, 5))

ax.fill_between(df['year'], df['co2_per_capita'], alpha=0.2, color='#7f8c8d')
ax.plot(df['year'], df['co2_per_capita'], 'o-', color='#2c3e50', linewidth=2.5, markersize=5)
ax.axvline(2016, color='red', linestyle='--', linewidth=1.5, label='Vision 2030 Launch')
ax.axhline(4.5, color='green', linestyle=':', linewidth=1.5, label='World Average (~4.5 MT)')

ax.set_title('CO₂ Emissions Per Capita (Metric Tons) — Saudi Arabia', fontweight='bold')
ax.set_xlabel('Year'); ax.set_ylabel('CO₂ per Capita (MT)')
ax.legend()
plt.savefig('q7_co2.png', dpi=150, bbox_inches='tight')
plt.show()

print(f"Saudi CO₂ per capita in 2023: {df.loc[df['year']==2023,'co2_per_capita'].values[0]} MT")
print(f"World average: ~4.5 MT")
print(f"Saudi Arabia emits roughly {df.loc[df['year']==2023,'co2_per_capita'].values[0]/4.5:.1f}x the world average")
```

**Answer Q7:** Saudi Arabia's CO₂ emissions per capita (~16.8 MT in 2023) remain roughly **3.7× the global average** of ~4.5 MT. While there's a very slight downward trend post-2016, it's not yet meaningful. Vision 2030 includes a target of generating 50% of electricity from renewables by 2030 — but this data suggests environmental transformation is the most lagging pillar of the plan.

### The Vision 2030 Verdict: Real Progress, but Oil Still Rules

After analyzing 33 years of Saudi Arabian development data, three main insights emerge:

**🟢 Clear Wins:**
- **Female workforce integration** is the most striking transformation. Female labor force participation nearly doubled from 17% to 34% between 2016 and 2023 — a generational shift driven by legislative reforms and new economic opportunities.
- **Tourism** reached a record 24.9M arrivals in 2023, with leisure tourism growing alongside the traditional Hajj/Umrah pilgrims, signaling early success of the Kingdom's entertainment and hospitality investments.

**🟡 Mixed Progress:**
- **Oil dependency is declining slowly**, but the 2022 GDP boom was still largely oil-price-driven. Non-oil sectors are growing in absolute terms, but the structural shift is gradual.
- **GDP per capita** hit $30K+ in 2022 but remains volatile — a Saudi citizen's economic reality still swings dramatically with oil markets.

**🔴 Lagging Areas:**
- **CO₂ emissions per capita** remain among the highest in the world and show no meaningful downward trend — the sustainability pillar of Vision 2030 needs much stronger action.
- **GDP growth volatility** persists. Without meaningful diversification, the economy will continue to boom and bust with oil prices.

**The Bottom Line:** Vision 2030 has produced real, measurable social transformation — particularly for women. Economic diversification is happening but slowly. The next decade will determine whether Saudi Arabia can reduce its oil dependency before the global energy transition reshapes the demand picture entirely.
